```python
numbers = [5, 4, 3, 6, 1, 9, 56, 2]
numbers.sort()
print("Sorted list:", numbers)

# Convert list to numpy array for using searchsorted
numbers_array = np.array(numbers)

# 2) Find the index and insert '7'
index_of_7 = np.searchsorted(numbers_array, 7)
#numbers_array[index_of_7]=7;# replace operation
numbers_array = np.insert(numbers_array, index_of_7, 7)
print("After inserting 7:", numbers_array)

# 3) While loop for user interaction
while True:
    user_input = input("Give a number (or type 'exit' to stop): ")
    if user_input.lower() == 'exit':
        break
    try:
        number = int(user_input)
        # 4) Find the index and insert the number using searchsorted
        index = np.searchsorted(numbers_array, number)
        numbers_array = np.insert(numbers_array, index, number) # dynamically growable array
        print(f"Array after inserting {number}:", numbers_array)
    except ValueError:
        print("Please enter a valid number.")
```

```python
categories = ['A', 'B', 'C', 'D', 'E']
values1 = [5, 7, 3, 8, 4]  # Group 1
values2 = [6, 5, 7, 6, 5]  # Group 2
values3 = [7, 4, 2, 3, 7] # Group 3

bar_width=0.2
x=np.arange(len(categories))
# x=1
# 1-0.2=0.8
plt.bar(x-bar_width/3, values1, width=bar_width,color='lightblue', label='Group 1')
plt.bar(x+bar_width/3, values2, width=bar_width,color='orange', label='Group 2', alpha=0.7)
plt.bar(x+bar_width, values3, width=bar_width,color='blue', label='Group 3', )
plt.plot(x-bar_width/3, values1,color='lightblue')
plt.plot(x+bar_width/3, values2,color='orange')
plt.plot(x+bar_width, values3,color='blue')
plt.xlabel('Categories')
plt.ylabel('Values')
plt.xticks(x, categories) # replace numeric values of x aaxis by the values of categories
plt.title('Bar Plot with Multiple Groups')
plt.legend(loc="upper right", bbox_to_anchor=(1.2,1))

plt.show()
```

```python
# Donut Chart
manufacturers=['Samsung','Apple','Huawei','Xiaomi','Others']
market_share=[30, 25, 18, 12, 15]
autopct_format=lambda p:f'{p*sum(market_share)/100:.1f}'
# 30*100/100= 30.0
# 25*100/100= 25.0
# 18*100/100= 18.0
# 12*100/100= 12.0
# 15*100/100= 15.0
plt.pie(market_share,labels=manufacturers,autopct=autopct_format,textprops={'fontsize':10,'fontweight':'bold', 'color':'black'})
circle=plt.Circle((0,0),0.6,color='white')
fig=plt.gcf()
fig.gca().add_artist(circle)
plt.title('Smartphone Market Contribution Share Analysis')
plt.show();
```

```python
# Sample student data
students = ['Jhon', 'Smith', 'Marry', 'Rose', 'Devid']
math_scores = [85, 92, 78, 88, 90]
science_scores = [76, 88, 92, 80, 78]
# Create a subplot with two subplots (1 row, 2 columns)
fig, axs = plt.subplots(1, 2, figsize=(12, 5))
# Plot Math Scores
axs[0].bar(students, math_scores, color='b')
axs[0].set_title('Math Scores')
axs[0].set_xlabel('Students')
axs[0].set_ylabel('Score')
# Plot Science Scores
axs[1].bar(students, science_scores, color='g')
axs[1].set_title('Science Scores')
axs[1].set_xlabel('Students')
axs[1].set_ylabel('Score')
# Adjust spacing between subplots
plt.tight_layout()
# Show the subplot
plt.show()
```

```python
# Sample student data
students = ['Jhon', 'Smith', 'Marry', 'Rose', 'Devid']
math_scores = [85, 92, 78, 88, 90]
science_scores = [76, 88, 92, 80, 78]
english_scores = [88, 96, 84, 90, 73]
cs_scores = [70, 85, 95, 85, 75]
# Create a subplot with two subplots (2 row, 2 columns)
fig, axs = plt.subplots(2, 2, figsize=(20, 8))
# Plot Math Scores
axs[0][0].bar(students, math_scores, color='b')
axs[0][0].set_title('Math Scores')
axs[0][0].set_xlabel('Students')
axs[0][0].set_ylabel('Score')
# Plot Science Scores
axs[0][1].bar(students, science_scores, color='g')
axs[0][1].set_title('Science Scores')
axs[0][1].set_xlabel('Students')
axs[0][1].set_ylabel('Score')
axs[1][0].bar(students, english_scores, color='red')
axs[1][0].set_title('English Scores')
axs[1][0].set_xlabel('Students')
axs[1][0].set_ylabel('Score')
axs[1][1].bar(students, cs_scores, color='orange')
axs[1][1].set_title('Computer Scores')
axs[1][1].set_xlabel('Students')
axs[1][1].set_ylabel('Score')
# Adjust spacing between subplots
plt.tight_layout()
# Show the subplot
plt.show()
```
